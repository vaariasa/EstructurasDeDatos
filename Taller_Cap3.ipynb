```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#@title 3.1.2

from IPython.display import Image, display
display(Image('/content/drive/MyDrive/CamScanner 03-13-2023 10.50_2.jpg', width=800))
```

```python
x=np.linspace(-10,10,1000)
r=3
x_dot= -np.cosh(x)+r
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')


plt.plot(x,x_dot)
plt.ylim(-10, 3)
plt.grid()

x=sp.symbols('x')
x_dot1= -sp.cosh(x)+r
roots=sp.solve(x_dot1)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')
```

```python
def fun (t,x,r,b):
  #return 1+r*x+x**b
  return -np.cosh(x)+r
  

r,b=3,0

ti=0; tf=1; dt=100
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-7,5,0.5):
  sol=solve_ivp(fun,tspam,[x0],args=(r,b),t_eval=t)
  plt.plot(sol.t, sol.y[0])


plt.axis([ti, tf, -10, 10])
plt.grid()
```

```python
ti=0; tf=100; dt=100
tspam=[ti, tf]

for r in np.arange(-1,5,0.2):
  for x0 in np.arange(-1,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(-1,5,0.2):
  for x0 in np.arange(-1,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([-1, 5, -5, 5])
plt.grid()
```
